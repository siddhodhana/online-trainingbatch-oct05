**matplotlib**
- one of the package in python.
- raw data converts into graphical visualization.
- histogram
- bar graph
- piechart
- box graph
- subplots
- linear graph
- scatter

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(10,100,20)
y=np.linspace(17,60,20)
plt.plot(x,y)
plt.show()
```

```python
x=[100,20,180,1]
y=[20,79,2,150]
a=[150,200,10,200]
b=[101,202,20,35]
c=[100,101,1,2,202]
d=[10,200,30,40,1]
plt.plot(x,y,"r-*",label="no of sixs")
plt.plot(a,b,"g-*",label="no of fours")
plt.plot(c,d,label="no of bounderies")
plt.legend()
plt.title("attendance list",color="red")
plt.xlabel("names",color="blue")
plt.ylabel("attendance",color="blue")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=[10,20,40,506,70]
y=np.log(x)
plt.figure(1)

plt.subplot(2,2,1)
plt.title("first subplot")
plt.xlabel("xlogx")
plt.ylabel("ylogx")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("ylog(x)**2")
#plt.xlabel("x_logx")
#plt.ylabel("y_logx**2")
plt.plot(x,y**2)

plt.figure(2)
plt.subplot(2,2,1)
plt.title("ylogx**3")
plt.xlabel("xlogx")
#plt.ylabel("")
plt.plot(x,y**3)

plt.subplot(2,2,2)
plt.title("fourth subplot")
plt.xlabel("xlogx")
#plt.ylabel("ylogx**4")
plt.plot(x,y)

plt.show()
```
